```python
from common.tools import copy_move, splicing, random_ps
import cv2
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/train_data.csv')

image = cv2.imread(df.iloc[1008]['image'])
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
image2 = df.iloc[1008]['similar']
mask = cv2.imread(df.iloc[1008]['mask'], cv2.IMREAD_GRAYSCALE) / 255
```

```python
f, ax = plt.subplots(1,3, figsize=(10,10))
ax[0].imshow(image);
ax[1].imshow(mask);
ax[2].imshow(image2);
```

```python
img, msk = copy_move(image, None, mask)
f, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(img);
ax[1].imshow(msk);
```

```python
img2, msk2 = copy_move(image, image2, mask)

f, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(img2);
ax[1].imshow(msk2);
```

```python
img2, msk2 = splicing(image, mask)

f, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(img2);
ax[1].imshow(msk2);
```

```python
img3, msk3 = random_ps(image, image2, mask)

f, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(img3);
ax[1].imshow(msk3);
```
